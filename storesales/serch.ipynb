```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('train.csv')
df_test=pd.read_csv('test.csv')
oil=pd.read_csv('oil.csv')
stores=pd.read_csv('stores.csv')
holiday=pd.read_csv('holidays_events.csv')
trans=pd.read_csv('transactions.csv')
```

```python
display(df.head())
display(oil.head())
display(stores.head())
display(holiday.head())
display(trans.head())
```

```python
df1=pd.merge(df,holiday,how='inner',on='date')
df2=pd.merge(df,holiday,how='left',on='date')
df4=pd.merge(df_test,holiday,how='inner',on='date')
df5=pd.merge(df_test,holiday,how='left',on='date')
```

```python
df1.info()
```

```python
df3=df2[df2.type.isna()]
df5=df5[df5.type.isna()]
```

```python
df3.drop(['type','locale','locale_name','description','transferred'],axis=1,inplace=True)
df5.drop(['type','locale','locale_name','description','transferred'],axis=1,inplace=True)
```

```python
display(df3.head())
df1.drop(['type','locale','locale_name','description','transferred'],axis=1,inplace=True)
df4.drop(['type','locale','locale_name','description','transferred'],axis=1,inplace=True)
```

```python
df1=pd.merge(df1,oil,how='left',on='date')
df3=pd.merge(df3,oil,how='left',on='date')
df1=pd.merge(df1,stores,how='left',on='store_nbr')
df3=pd.merge(df3,stores,how='left',on='store_nbr')
df1=pd.merge(df1,trans,how='left',on=['date','store_nbr'])
df3=pd.merge(df3,trans,how='left',on=['date','store_nbr'])
```

```python
df4=pd.merge(df4,oil,how='left',on='date')
df5=pd.merge(df5,oil,how='left',on='date')
df4=pd.merge(df4,stores,how='left',on='store_nbr')
df5=pd.merge(df5,stores,how='left',on='store_nbr')
df4=pd.merge(df4,trans,how='left',on=['date','store_nbr'])
df5=pd.merge(df5,trans,how='left',on=['date','store_nbr'])
```

```python
le=LabelEncoder()
df1['familynum']=le.fit_transform(df1.family)
df3['familynum']=le.fit_transform(df3.family)
df4['familynum']=le.fit_transform(df4.family)
df5['familynum']=le.fit_transform(df5.family)
```

```python
trainforna1=df1[['store_nbr','onpromotion','familynum','dcoilwtico','cluster','transactions','sales']].fillna(0)
trainforna2=df1[['store_nbr','onpromotion','familynum','dcoilwtico','sales']].fillna(0)
trainfornotna1=df3[['store_nbr','onpromotion','familynum','dcoilwtico','cluster','transactions','sales']].fillna(0)
trainfornotna2=df3[['store_nbr','onpromotion','familynum','dcoilwtico','sales']].fillna(0)

testforna1=df4[['store_nbr','onpromotion','familynum','dcoilwtico','cluster','transactions']].fillna(0)
testforna2=df4[['store_nbr','onpromotion','familynum','dcoilwtico']].fillna(0)
testfornotna1=df5[['store_nbr','onpromotion','familynum','dcoilwtico','cluster','transactions']].fillna(0)
testfornotna2=df5[['store_nbr','onpromotion','familynum','dcoilwtico']].fillna(0)
```

```python
xna1=trainforna1.drop('sales',axis=1)
yna1=trainforna1[['sales']]
xna2=trainforna2.drop('sales',axis=1)
yna2=trainforna2[['sales']]
datas=((xna1,yna1,testforna1),(xna2,yna2,testforna2))
```

```python
#holiday以外
#dfs=[]
for data in datas:
    xna=data[0]
    yna=data[1]
    test=data[2]
    x_trainna,x_testna,y_trainna,y_testna=train_test_split(xna,yna,test_size=0.25,random_state=0)
    from sklearn.ensemble import RandomForestRegressor
    scores=[]
    model=RandomForestRegressor(random_state=0)
    model.fit(x_trainna,y_trainna)
    tmp=model.predict(test)
    test['sales']=tmp
    #dfs.append(tmp)
    sc_trna=model.score(x_trainna,y_trainna)
    sc_tena=model.score(x_testna,y_testna)
    print(sc_trna,sc_tena)
```
